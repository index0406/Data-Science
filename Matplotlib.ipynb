# Matplotlib
### Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoints=np.array([1,2])
ypoints=np.array([5,10])
plt.plot(xpoints,ypoints)
```

```python
# if x points are not given then it is 1,2,3
y=np.array([3,8,12,9,15])
plt.plot(y,marker='*')
plt.show()
```

```python
plt.plot(ypoints,marker='X')
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints, '*'':r')
```

```python
xpoints=np.array([1,6])
ypoints=np.array([4,5])
plt.plot(xpoints,ypoints,'+')
```

```python
ypoints=np.array([3,8,1,10])
xpoints=([5,6,7,9])
plt.plot(xpoints, ypoints, marker='o', ms=2, color='green')
```

```python
plt.plot(xpoints, ypoints, marker='o', ms=20, color='green')
```

###### ms is marker size
###### mec is marker edge color
###### mfc is for marker face color

```python
ypoints=([3,8,1,10])
plt.plot(ypoints, marker='*', ms=205, mec='red')
```

```python
ypoints=np.array([6,2,9,12])
plt.plot(ypoints, marker='o', ms=100, mec='black', mfc='red')
```

```python
ypoints=np.array([6,2,9,12])
plt.plot(ypoints, linestyle='dashed', marker='o', ms=20, mec='hotpink', mfc='blue')
```

```python
ypoints=np.array([6,2,9,12])
plt.plot(ypoints, linestyle='dashed', marker='o', ms=20, mec='hotpink', mfc='blue', color='black', linewidth=6)
```

```python
x=np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y=np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.plot(x, y, color='r', linewidth=1)
plt.xlabel("Average Pulse", fontsize=10, color="red")
plt.ylabel("Calories Burnage", fontsize=10, color="green")
```

```python
x=np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y=np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.plot(x, y)
plt.title("Sports Watch Data", loc="left", fontsize=20, color="green")
plt.xlabel("Average Pulse", loc="left")
plt.ylabel("Calories Burnage")
```

```python
x=np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y=np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.plot(x, y)
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calories Burnage")
plt.grid()
```

```python
x=np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y=np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.plot(x, y, linewidth=5, color="black")
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calories Burnage")
plt.grid(color="green", linestyle="--", linewidth=2.5)
```

```python
x=np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6])
y=np.array([99, 86, 87, 88, 111, 86, 103, 81, 94, 78, 77, 85, 86])
plt.scatter(x,y)
```

```python
x=np.array([1, 2, 3, 4, 5])
y=np.array([6, 7, 8, 9, 10])
colors=(["red", "green", "pink", "black", "blue"])
size=([10, 30, 50, 70, 80])
plt.scatter(x, y, c=colors, s=size)
```

```python
x=np.array(["A", "B", "C", "D"])
y=np.array([3, 8, 1, 10])
plt.bar(x,y)
```

```python
x=np.array(["A", "B", "C", "D"])
y=np.array([3, 8, 1, 10])
colors=(["red", "black", "green", "blue"])
plt.bar(x, y, color=colors)
```

```python
x=np.array(["A", "B", "C", "D"])
y=np.array([3, 8, 1, 10])
plt.barh(x, y, height=0.5)
```

```python
x=np.array(["A", "B", "C", "D"])
y=np.array([3, 8, 1, 10])
plt.barh(x, y, height=0.9)
```

```python
y=np.array([10, 135, 25, 15, 19])
plt.pie(y)
plt.show()
```

```python
x=np.array([18, 25, 15, 10])
plt.pie(y, startangle=90)
plt.show()
```

```python
x=np.array([18, 25, 15, 10])
mylabels = ["Cherries","Apples","Bananas","Dates"]
plt.pie(x, labels=mylabels)
plt.show()
```

```python
x=np.array([18, 25, 15, 10])
mylabels = ["Cherries","Apples","Bananas","Dates"]
mycolors = ["black", "blue", "pink", "red"]
plt.pie(x, labels=mylabels, colors=mycolors)
plt.show()
```

```python
x=np.array([18, 25, 15, 10])
mylabels = ["Cherries","Apples","Bananas","Dates"]
mycolors = ["black", "blue", "pink", "red"]
plt.pie(x, colors=mycolors)
plt.legend(labels=mylabels)
plt.show()
```

```python
x=np.array([18, 25, 15, 10])
mylabels = ["Cherries","Apples","Bananas","Dates"]
myexplode = [0, 0.95, 0.1, 3.2]
plt.pie(x, labels=mylabels, explode=myexplode)
plt.show()
```
